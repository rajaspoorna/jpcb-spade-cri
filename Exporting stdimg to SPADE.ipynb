```python
UpMasks = ['UpR01', 'UpR02', 'UpR03', 'UpR04', 'UpR05', 'UpR06', 'UpR07', 'UpR08', 'UpR09', 'UpR10']
plt.figure(figsize=(12,6), dpi=300)
UpGonadMean = ct.stdimg[masks['UpGonad']].mean(axis=0)[fingerprint_start:fingerprint_end]
for name in UpMasks:
    plt.plot(ct.stdfreq_crop,
             ct.stdimg[masks[name]].mean(axis=0)[fingerprint_start:fingerprint_end],# - UpGonadMean,
             label=name,
             lw=1)
plt.legend()
plt.show()
```
